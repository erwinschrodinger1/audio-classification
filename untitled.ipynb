# Audio Source Seperation
A basic start-up to get started with DSP (Digital Signal Processing)

### Importing Necessary Libaries 

```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Loading audio file and displaying it

```python
signal, rate = librosa.load('mixkit-melodic-guitar-tones-2341.wav')
librosa.display.waveshow(signal,sr=rate)
```

### Short Fourier Transfor with default window size

```python
sft_audio = librosa.stft(signal)
plt.plot(sft_audio)
```

### Spectorgram for the given fft signal

```python
spectorgram = np.abs(sft_audio)
librosa.display.specshow(spectorgram,sr=rate,x_axis='time',y_axis='hz')
plt.colorbar()
```

### Melody Spectogram

```python
mel_spectorgram = librosa.feature.melspectrogram(y=signal,sr=rate)
librosa.display.specshow(mel_spectorgram,sr=rate,x_axis='time',y_axis='mel')
plt.colorbar()
```

### Plotting the spectogram in log scale

```python
log_mel_spectorgram = librosa.power_to_db(mel_spectorgram)
librosa.display.specshow(log_mel_spectorgram,sr=rate,x_axis='time',y_axis='mel')
plt.colorbar()
```

### MFCC (Mel Frequency Cepstral Coefficients)

```python
cepstrum = librosa.feature.mfcc(y=signal,sr=rate)
librosa.display.specshow(cepstrum,sr=rate,x_axis='time',y_axis='hz')
plt.colorbar()
```

### Removing quiet audio and replotting everything

```python
window_size = int(rate/100) 
threshold = 0.00001

rolling_mean = pd.Series(signal).rolling(window=window_size,min_periods=1, center=True).mean()


signal_updated = signal[rolling_mean > threshold]
plt.plot(signal[:], color='blue')
plt.plot(signal_updated[:],color='orange')
signal = signal_updated
plt.show()
```

```python
sft_audio = librosa.stft(signal)
plt.plot(sft_audio)
```

```python
spectorgram = np.abs(sft_audio)
librosa.display.specshow(spectorgram,sr=rate,x_axis='time',y_axis='hz')
plt.colorbar()
```

```python
mel_spectorgram = librosa.feature.melspectrogram(y=signal,sr=rate)
librosa.display.specshow(mel_spectorgram,sr=rate,x_axis='time',y_axis='mel')
plt.colorbar()
```

```python
log_mel_spectorgram = librosa.power_to_db(mel_spectorgram)
librosa.display.specshow(log_mel_spectorgram,sr=rate,x_axis='time',y_axis='mel')
plt.colorbar()
```

```python
cepstrum = librosa.feature.mfcc(y=signal,sr=rate)
librosa.display.specshow(cepstrum,sr=rate,x_axis='time',y_axis='hz')
plt.colorbar()
```
